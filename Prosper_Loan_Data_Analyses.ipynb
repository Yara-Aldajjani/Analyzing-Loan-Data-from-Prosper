### Data Cleaning
1. Fill in the null values of occupation and DebtToIncomeRatio
2. Pick out features of intrest
3. Drop the rows with ListingNumber duplication
4. Change `TotalTrades and TotalInquiries` datatypes to Integer, and `ListingCategory (numeric)` to Category
5. Drop rows with no ProsperRating data
6. Separate the time, day, month and year into four columns from `ListingCreationDate` column and give names to the months

```python
# Pick out features of intrest
#Code:
features_intrest= ['ListingNumber','ListingCreationDate','LoanOriginalAmount','LoanStatus','ListingCategory (numeric)','BorrowerState','BorrowerAPR','BorrowerRate','StatedMonthlyIncome', 'ProsperRating (Alpha)', 
        'Occupation','Term','EmploymentStatus','TotalInquiries','DebtToIncomeRatio','MonthlyLoanPayment','TotalTrades','Investors']
Loan_Data_subset= Loan_Data[features_intrest]
```

```python
#Test: data information
Loan_Data_subset.info()
```

```python
#Test: sample of the data
Loan_Data_subset.sample(15)
```

```python
# Drop the rows with ListingNumber duplication  
#Code:
Loan_Data_subset.drop_duplicates(inplace=True)
```

```python
#Test: data information
Loan_Data_subset.info()
```

```python
#Test: Check for duplication
Loan_Data_subset.duplicated().any()
```

```python
#Drop rows with no ProsperRating data
#Code:
Loan_Data_subset= Loan_Data_subset[Loan_Data_subset['ProsperRating (Alpha)'].notnull()]
```

```python
#Test: data information
Loan_Data_subset.info()
```

```python
#Test: Check for null values
Loan_Data_subset['ProsperRating (Alpha)'].isnull().any()
```

```python
#Fill in the null values of occupation and DebtToIncomeRatio
#Code 1:
Loan_Data_subset['Occupation'] = Loan_Data_subset['Occupation'].fillna('Unknown')
```

```python
#Code 2:
Loan_Data_subset['DebtToIncomeRatio'] = Loan_Data_subset['DebtToIncomeRatio'].fillna(Loan_Data_subset['DebtToIncomeRatio'].mean())
```

```python
# Test: data information
Loan_Data_subset.info()
```

```python
# Test 1: Check for null values in Occupation
Loan_Data_subset['Occupation'].isnull().any()
```

```python
# Test 2: Check for null values in DebtToIncomeRatio
Loan_Data_subset['DebtToIncomeRatio'].isnull().any()
```

```python
#Change TotalTrades and TotalInquiries datatypes to Integer, and ListingCategory (numeric) to Category
#Code 1:
Loan_Data_subset['TotalInquiries'] = Loan_Data_subset['TotalInquiries'].astype(int)
Loan_Data_subset['TotalTrades'] = Loan_Data_subset['TotalTrades'].astype(int)
```

```python
# Code 2:
Loan_Data_subset['ListingCategory (numeric)'] = Loan_Data_subset['ListingCategory (numeric)'].astype("category")
```

```python
# Test: data information
Loan_Data_subset.info()
```

```python
# Test 1 and 2: Check for TotalTrades, TotalInquiries and ListingCategory (numeric) datatypes
print(Loan_Data_subset['TotalInquiries'].dtypes)
print(Loan_Data_subset['TotalTrades'].dtypes)
print(Loan_Data_subset['ListingCategory (numeric)'].dtypes)

```

```python
# Separate the time, day, month and year into four columns from `ListingCreationDate` column and give names to the months
#Code 1: Splitting into year
Loan_Data_subset['Year']=Loan_Data_subset['ListingCreationDate'].apply(lambda x: x.split("-")[0]).astype(str)
```

```python
#Test 1: sample of the Year column
Loan_Data_subset['Year'].sample(10)
```

```python
#Code 2.1: Splitting into month
Loan_Data_subset['Month'] = Loan_Data_subset['ListingCreationDate'].apply(lambda x: x.split("-")[1]).astype(str)
```

```python
#Test 2.1: sample of the Month column
Loan_Data_subset['Month'].sample(10)
```

```python
#Code 2.2: give names to the months instead of the numbers
Loan_Data_subset['Month'].replace(['01','02','03','04','05','06','07','08','09','10','11','12'],['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec'],inplace = True)
```

```python
#Test 2.2: sample of the Month column
Loan_Data_subset['Month'].sample(10)
```

```python
#Code:
Loan_Data_subset['ListingCreationDate'] = Loan_Data_subset['ListingCreationDate'].apply(lambda x: x.split("-")[2]).astype(str)
```

```python
#Code 3:
Loan_Data_subset['Day'] = Loan_Data_subset['ListingCreationDate'].apply(lambda x: x.split(" ")[0]).astype(str)
```

```python
#Test 3: sample of Day column
Loan_Data_subset['Day'].sample(5)
```

```python
#Code 4:
Loan_Data_subset['Time'] = Loan_Data_subset['ListingCreationDate'].apply(lambda x: x.split(" ")[1]).astype(str)
```

```python
#Test 4: sample of Time column
Loan_Data_subset['Time'].sample(5)
```

```python
# Code: Dropping the ListingCreationDate column
Loan_Data_subset.drop(columns = ['ListingCreationDate'],inplace = True)
```

```python
#Test: data information
Loan_Data_subset.info()
```
